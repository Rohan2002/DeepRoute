```python
parent_directory = os.path.dirname(os.getcwd())
image_name = "parent_PrL-4-48e876cc-0ab9-4230-8c31-2b39c10bc662.jpg"
image_path = f"{parent_directory}/data/train/{image_name}"
```

```python
def show_img(img_arr):
    plt.figure(figsize=(20, 20))
    plt.imshow(img_arr)
    plt.show()


def clear_img(img_path: str):
    img_arr = cv2.imread(img_path)
    kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
    img_sharp = cv2.filter2D(img_arr, -1, kernel)
    return img_sharp
```

```python
img_sharp = clear_img(image_path)
show_img(img_sharp)
```
